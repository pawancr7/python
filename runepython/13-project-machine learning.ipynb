```python
import numpy as np 
import csv 
from sklearn.linear_model  import LinearRegression
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
with open ("data/cars.csv","r") as f:
    csv_reader = csv.DictReader(f)
    cars = list(csv_reader)
    
```

```python
cars
```

```python
horsepower = [int (car['Horsepower'] )for car in cars ]
torque = [int (car ['Torque'] ) for car in cars ]
```

```python
horsepower 
```

```python
torque
```

```python
len(horsepower)
```

```python
X  = np.array(horsepower).reshape((-1,1))
Y  = np.array(torque).reshape((-1,1))
```

```python
X
```

```python
Y
```

## apply linear regression 

```python
lin_regresion = LinearRegression ()
lin_regresion.fit(X,Y)
Y_pred = lin_regresion.predict(X)
```

```python
alpha = str(round (lin_regresion.intercept_[0],5))
beta = str(round(lin_regresion.coef_[0][0]))
```

```python

fig ,ax = plt.subplots()

ax.set_title(f"Alpha{alpha} Beta {beta}")
ax.scatter(X,Y)
ax.plot(X,Y_pred,c='r')
```
