```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
plt.plot([1,2,3,4,5])
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4])
```

```python
x = [1,2,3,4,5,6]
y = [2,3,4,1,5,3]
c = [2,3,5,7,1,9]
fig, ax = plt.subplots()

ax.scatter(x,y,c=c)
ax.set_title("Title")
ax.set_xlabel("X label")
ax.set_ylabel("y label")

```

```python
data = [1,1,2,2,3,4,5,2,2,1,4,5,6,2,3]

fig,ax = plt.subplots()
ax.hist(data,bins=5)
ax.set_title("tittle")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
```
