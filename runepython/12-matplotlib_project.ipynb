## car horsepower , torque and fuel mpg 

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
import csv
```

```python
with open("data/cars.csv","r") as f :
    csv_reader = csv.DictReader(f)
    cars = list(csv_reader)
```

```python

cars[0]
```

```python
horsepower = [int (car['Horsepower']) for car in cars ]
```

```python
horsepower
```

```python
torque = [int(car['Torque']) for car in cars]
```

```python
torque
```

```python
fuelmpg = [int(car['Fuel City mpg']) for car in cars ]
```

```python
fuelmpg
```

```python
fig ,ax = plt.subplots()
ax.scatter(horsepower,torque,alpha=.1,c=fuelmpg)
ax.set_xlabel("horsepower")
ax.set_ylabel("torque")
```

## Roll Dice

```python
import random
```

```python
def roll_dice():
    return random.randint(1,6) + random.randint(1,6)
```

```python
roll_dice()
```

```python
data = [roll_dice() for i in range(500)]
```

```python
len(data)
```

```python
max(data)
```

```python
min(data)
```

```python
fig ,ax = plt.subplots()
ax.hist(data,bins = 11 )
```
